```python
order = pd.concat([df['order_id'],df['order_date'],df['order_priority'],df['ship_date'],df['ship_mode'],df['shipping_cost'],df['sales'],df['quantity'],df['discount'],df['profit'],df['market']], axis=1).drop_duplicates(subset=['order_id'])
order
```

```python
customer_df = df[['customer_name', 'segment']].drop_duplicates().reset_index(drop=True)
customer_df['customer_id'] = customer_df.index
customer_df
```

```python
location_df = df[['state', 'country', 'region']].drop_duplicates().reset_index(drop=True)
location_df['location_id'] = location_df.index
location_df
```

```python
category_df = df[['category', 'sub_category']].drop_duplicates().reset_index(drop=True)
category_df['category_id'] = category_df.index
category_df
```

```python
product_df=pd.merge(df, category_df, on=['category', 'sub_category'])
product_df = product_df[['product_name','category_id']].drop_duplicates().reset_index(drop=True)
product_df['product_id'] = product_df.index
product_df
```

```python
order_df=pd.merge(df, customer_df, on=['customer_name', 'segment'])
order_df=pd.merge(order_df, location_df, on=['state', 'country','region'])
order_df=pd.merge(df, category_df, on=['category', 'sub_category'])
order_df=pd.merge(order_df, product_df, on=['product_name','category_id'])
order_df = order_df[['order_id','order_date','order_priority','ship_date','ship_mode','shipping_cost','sales','quantity','discount','profit','market']].drop_duplicates().reset_index(drop=True)
order_df
```

2. Izraditi bazu podataka u DBMS-u (logički dizajn baze, npr. u MySQL)

3. Napuniti bazu podataka sa podacima iz CSVa
- napisati skriptu u Pythonu koja će napuniti bazu

```python
import psycopg2
from sqlalchemy import create_engine

conn_string = 'postgresql://postgres:admin@127.0.0.1/spi-projekt'

db = create_engine(conn_string)

try:
    order_table = order_df.to_sql('orders', db, if_exists='replace', index=False)
    customer_table = customer_df.to_sql('customers', db, if_exists='replace', index=False)
    location_table = location_df.to_sql('locations', db, if_exists='replace', index=False)
    product_table = product_df.to_sql('products', db, if_exists='replace', index=False)
    category_table = category_df.to_sql('categories', db, if_exists='replace', index=False)
except Exception as ex:
    print(ex)
else:
    print("PostgreSQL tablice su kreirane i popunjene.")
```
