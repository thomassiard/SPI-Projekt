## Izrada relacijskog modela i baze podataka

1. Izraditi ER dijagram (konceptualni dizajn baze):
- identificirati entitete
- identificirati atribute entiteta
- definirati odnose između entiteta
- definirati kardinalnost

![ER Dijagram](ER_Dijagram.png "ER Dijagram")

```python
df[['region_id']] = df[['region']].apply(lambda x: pd.factorize(x)[0])
df[['state_id']] = df[['state']].apply(lambda x: pd.factorize(x)[0])
df[['customer_id']] = df[['customer_name']].apply(lambda x: pd.factorize(x)[0])
df['product_id'] = df[['product_id']].apply(lambda x: pd.factorize(x)[0])
df[['order_priority_id']] = df[['order_priority']].apply(lambda x: pd.factorize(x)[0])
df[['ship_mode_id']] = df[['ship_mode']].apply(lambda x: pd.factorize(x)[0])
df['country_id'] = df['country']
df['sub_category_id'] = df['sub_category']
df['category_id'] = df['category']
df['sub_category_name'] = df['sub_category']
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
state_df = df[['state', 'country', 'region']].drop_duplicates(subset=['state'], keep='first').reset_index(drop=True)
state_df[['state_id', 'country_id', 'region_id']] = state_df[['state', 'country', 'region']].apply(lambda x: pd.factorize(x)[0])
state_df
```

```python
country_df = state_df[['country', 'country_id', 'region_id']].drop_duplicates().reset_index(drop=True)
country_df
```

```python
region_df = state_df[['region', 'region_id']].drop_duplicates().reset_index(drop=True)
region_df
```

```python
state_df = state_df[['state', 'state_id', 'country_id']].drop_duplicates().reset_index(drop=True)
state_df
```

```python
product_df = df[['product_name','product_id','sub_category_id']].drop_duplicates().reset_index(drop=True)
product_df['product_id'] = product_df[['product_id']].apply(lambda x: pd.factorize(x)[0])
product_df['sub_category_id'] = product_df[['sub_category_id']].apply(lambda x: pd.factorize(x)[0])
product_df
```

```python
customer_df = df[['customer_name', 'segment']].drop_duplicates().reset_index(drop=True)
customer_df['customer_id'] = customer_df.index
customer_df
```

```python
category_df = df[['category', 'category_id']].drop_duplicates().reset_index(drop=True)
category_df['category_id'] = category_df.index
category_df
```

```python
sub_category_df = df[['sub_category_name','category_id','sub_category_id']].drop_duplicates().reset_index(drop=True)
sub_category_df['sub_category_id'] = sub_category_df[['sub_category_id']].apply(lambda x: pd.factorize(x)[0])
sub_category_df['category_id'] = sub_category_df[['category_id']].apply(lambda x: pd.factorize(x)[0])
sub_category_df
```

```python
ship_mode_df = df[['ship_mode']].drop_duplicates().reset_index(drop=True)
ship_mode_df['ship_mode_id'] = ship_mode_df.index
ship_mode_df
```

```python
order_priority_df = df[['order_priority']].drop_duplicates().reset_index(drop=True)
order_priority_df['order_priority_id'] = order_priority_df.index
order_priority_df
```

2. Izraditi bazu podataka u DBMS-u (logički dizajn baze, npr. u MySQL)

3. Napuniti bazu podataka sa podacima iz CSVa
- napisati skriptu u Pythonu koja će napuniti bazu

```python
order_df = df[['order_id','order_date','order_priority_id','ship_date','ship_mode_id','shipping_cost','sales','quantity','discount','profit','state_id','customer_id','product_id']].drop_duplicates(subset=['order_id'])
order_df
```

```python
import psycopg2
from sqlalchemy import create_engine

conn_string = 'postgresql://postgres:admin@127.0.0.1/spi-projekt'

db = create_engine(conn_string)

try:
    order_table = order_df.to_sql('orders', db, if_exists='replace', index=False)
    customer_table = customer_df.to_sql('customers', db, if_exists='replace', index=False)
    state_table = state_df.to_sql('states', db, if_exists='replace', index=False)
    product_table = product_df.to_sql('products', db, if_exists='replace', index=False)
    category_table = category_df.to_sql('categories', db, if_exists='replace', index=False)
    sub_categoryy_table = sub_category_df.to_sql('sub_categories', db, if_exists='replace', index=False)
    country_table = country_df.to_sql('countries', db, if_exists='replace', index=False)
    region_table = region_df.to_sql('regions', db, if_exists='replace', index=False)
    ship_mode_table = ship_mode_df.to_sql('ship_modes', db, if_exists='replace', index=False)
    order_priority_table = order_priority_df.to_sql('order_priorities', db, if_exists='replace', index=False)
except Exception as ex:
    print(ex)
else:
    print("PostgreSQL tablice su kreirane i popunjene.")
```

## 2. ETL - napuniti skladište podataka sa podacima

- koristiti alat Pentaho Data Integration
- Upute za  korištenje Pentaho: Upute 1, Upute 2

```python
import numpy as np
```

```python
dim_location_df = pd.merge(country_df, region_df, on="region_id")
dim_location_df = dim_location_df.merge(state_df, on="country_id")
dim_location_df = dim_location_df[['region', 'country', 'state', 'state_id']]
dim_location_df['dim_location_id'] = dim_location_df.index
dim_location_df['start_date'] = pd.Timestamp.today()
dim_location_df['end_date'] = np.nan
dim_location_df
```

```python
dim_time_df = order_df['order_date'].unique()
dim_time_df = pd.DataFrame({'order_date': pd.to_datetime(dim_time_df, format='%d-%m-%Y')})
dim_time_df['dim_time_id'] = dim_time_df.index + 1
dim_time_df['year'] = dim_time_df['order_date'].dt.year
dim_time_df['quarter'] = dim_time_df['order_date'].dt.quarter
dim_time_df['month'] = dim_time_df['order_date'].dt.month
dim_time_df['day'] = dim_time_df['order_date'].dt.day
dim_time_df
```

```python
categories_df = pd.merge(category_df, sub_category_df, on="category_id")
dim_products_df = product_df.merge(categories_df, on="sub_category_id")
dim_products_df = dim_products_df[['category', 'sub_category_name', 'product_name', 'product_id']]
dim_products_df['dim_product_id'] = dim_products_df.index
dim_products_df['start_date'] = pd.Timestamp.today()
dim_products_df['end_date'] = np.nan
dim_products_df
```

```python
dim_customer_df = customer_df.copy()
dim_customer_df['dim_customer_id'] = dim_customer_df.index
dim_customer_df['start_date'] = pd.Timestamp.today()
dim_customer_df['end_date'] = np.nan
dim_customer_df
```

```python
dim_order_information_df = pd.merge(order_df, ship_mode_df, on="ship_mode_id")
dim_order_information_df = dim_order_information_df.merge(order_priority_df, on="order_priority_id")
dim_order_information_df = dim_order_information_df[['order_priority', 'discount', 'profit', 'shipping_cost', 'ship_mode', 'order_id']]
dim_order_information_df['dim_order_inf_id'] = dim_order_information_df.index
dim_order_information_df['start_date'] = pd.Timestamp.today()
dim_order_information_df['end_date'] = np.nan
dim_order_information_df
```

```python
dim_order_df = pd.merge(order_df, dim_location_df, on = "state_id")
dim_order_df = dim_order_df.drop(['start_date', 'end_date'], axis = 1)
dim_order_df = dim_order_df.merge(dim_time_df, on = "order_date")
dim_order_df = dim_order_df.merge(dim_products_df, on = "product_id")
dim_order_df = dim_order_df.drop(['start_date', 'end_date'], axis = 1)
dim_order_df = dim_order_df.merge(dim_customer_df, on = "customer_id")
dim_order_df = dim_order_df.drop(['start_date', 'end_date'], axis = 1)
dim_order_df = dim_order_df.merge(dim_order_information_df, on = "order_id")
dim_order_df = dim_order_df[['sales', 'quantity', 'dim_customer_id', 'dim_product_id', 'dim_location_id', 'dim_time_id', 'dim_order_inf_id', 'order_id']]
dim_order_df['ft_order_id'] = dim_order_df.index
dim_order_df
```

```python
conn_string = 'postgresql://postgres:admin@127.0.0.1/SPI-DIM-projekt'

db = create_engine(conn_string)

try:
    dim_location_table = dim_location_df.to_sql('dim_locations', db, if_exists='replace', index=False)
    dim_time_table = dim_time_df.to_sql('dim_time', db, if_exists='replace', index=False)
    dim_products_table = dim_products_df.to_sql('dim_products', db, if_exists='replace', index=False)
    dim_customer_table = dim_customer_df.to_sql('dim_customers', db, if_exists='replace', index=False)
    dim_order_information_table = dim_order_information_df.to_sql('dim_order_informations', db, if_exists='replace', index=False)
    dim_order_table = dim_order_df.to_sql('dim_orders', db, if_exists='replace', index=False)
except Exception as ex:
    print(ex)
else:
    print("PostgreSQL tablice su kreirane i popunjene.")
```
